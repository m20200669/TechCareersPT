```python
import plotly.io as pio
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
import numpy as np
import plotly.figure_factory as ff
```

```python
df = pd.read_excel('https://github.com/vbussolaro/Tech-Career-Report-2021/raw/main/Tech%20Careers%20Report%20PT%202021%20-%20Raw%20Data.xlsx', sheet_name='TCS PT 2021 - raw data')
```

```python
df
```

```python
fig = ff.create_distplot([df[df['Gender']=='Male']['Age'], df[df['Gender']=='Female']['Age']], 
                         ['Male', 'Female'],
                         colors = ['#2BCDC1', '#F66095'],
                         bin_size = [2, 2],
                         show_rug=False, #rug 
                         show_hist=False, #hist
                         show_curve=True # curve
                        )
fig.show()
```

```python
fig = ff.create_distplot([df[df['Gender']=='Male']['Age'], df[df['Gender']=='Female']['Age']], 
                         ['Male', 'Female'],
                         colors = ['#2BCDC1', '#F66095'],
                         bin_size = [2, 2]
                        )
fig.show()
```

```python
#type 1 : Stacked graph (default)

fig = px.histogram(df, x="Avg_Salary", y="Gender", color="Citizenship")
fig.show()

#type 2 : group
fig = px.histogram(df, y="Avg_Salary", x="Working_Experience_Aggregated", color="Working_Experience_Aggregated")
fig.show()
```

```python
fig = px.scatter(df, x='Citizenship', y='Avg_Salary', opacity=0.5)
fig.show()
```

```python
# DF without NA on the selected columns to perform sunburst-treemap
df2 = df.dropna(axis=0, how='any', thresh=None, subset=['Way_Into_Tech', 'Education_Level','Work_Company_Country'])
```

```python
fig = px.treemap(df2, path=['Job_Remote_or_Office','Work_Company_Country'], values='Avg_Salary')
fig.show()
```

```python
fig = px.sunburst(df2, path=['Way_Into_Tech','Education_Level'], values='Avg_Salary')
fig.show()
```

```python
fig = px.treemap(df2, path=['Way_Into_Tech','Education_Level',], values='Avg_Salary')
fig.show()
```
